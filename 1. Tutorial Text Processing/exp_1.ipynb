```python
import os
import nltk
nltk.data.path.append(os.path.join(os.getcwd(), "nltk_data"))
import re
```

```python
# Another sample text
text = "Dr. Smith graduated from the University of Washington. He later started an analytics firm called Lux, which catered to enterprise customers."
print(text)
```

```python
from nltk.tokenize import word_tokenize

w_tokens = word_tokenize(text)
print(f'type of w_tokens : {type(w_tokens)}')
print(f'w_tokens : {w_tokens}')
```

```python
from nltk.tokenize import sent_tokenize

s_tokens = sent_tokenize(text)
print(f'type of s_tokens : {type(s_tokens)}')
print(f's_tokens : {s_tokens}')
```

```python
# stopwords
from nltk.corpus import stopwords
print(f'len of stopwords : {len(stopwords.words("english"))}')
```

```python
# Reset text
text = "The first time you see The Second Renaissance it may look boring. Look at it at least twice and definitely watch part 2. It will change your view of the matrix. Are the human people the ones who started the war ? Is AI a bad thing ?"

# normalize
norm_text = re.sub(r'[^a-zA-Z0-9]', ' ', text.lower())

words = norm_text.split()

print(f'after normalizing : {words}')
```

```python
# removal of stop words
rem_stop_wrds_text = [w for w in words if w not in stopwords.words('english')]
print(f'After removal of stopwords : {rem_stop_wrds_text}')
```

```python
# Stemming
from nltk.stem.porter import PorterStemmer

stemmed = [PorterStemmer().stem(w) for w in rem_stop_wrds_text]
print(f'After running porter stemmer : {stemmed}')
```

```python
# Lemmitization
from nltk.stem.wordnet import WordNetLemmatizer

lemma = [WordNetLemmatizer().lemmatize(w) for w in rem_stop_wrds_text]
print(f'After running lemmatizer : {lemma}')
```
